### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
DataFrame1 = pd.merge(recipes[['id','name']], reviews[['recipe_id','user_id','rating']], left_on='id', right_on='recipe_id')
DataFrame1.drop("recipe_id", axis=1, inplace=True)

print("Полученный DataFrame:\n", DataFrame1)

print("\nПроверка. \nРецепт с id 276594 не имеет отзывов.")
print(DataFrame1[DataFrame1.id == 276594])
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
Data1 = recipes[['id', 'name']]

Data2 = reviews.groupby('recipe_id')[['user_id']].count()

Data2.columns = ['review_count']

DataFrame2 = pd.merge(Data1, Data2, left_on='id', right_on='recipe_id', how='left')

print("Список рецептов не имеющих отзывов:")
print(DataFrame2[DataFrame2.review_count.isna()])

DataFrame2['review_count'] = (DataFrame2['review_count'].fillna(0.0)).astype('int64')

print("\nПолученный DataFrame:")
print(DataFrame2)

print("\nПроверка. \nРецепт с id 276594 не имеет отзывов.")
print(DataFrame2[DataFrame2.id == 276594])
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
idDate = recipes[['id','submitted']]
idRating = reviews[['recipe_id','rating']]
idDateRating = idRating.merge(idDate, left_on='recipe_id', right_on='id', how='left') # 
meanYears = idDateRating.groupby(idDateRating.submitted.dt.year)['rating'].mean()

print("Год в котором добавленные рецепты имеют наименьший средний рейтинг:")
meanYears[meanYears == meanYears.min()].index[0]
```
